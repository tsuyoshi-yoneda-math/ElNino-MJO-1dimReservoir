```python
import requests
import pandas as pd
import io

url = "https://ds.data.jma.go.jp/tcc/tcc/products/clisys/mjo/figs/olr0-sst1_1991-2021/rmm8.csv"
res = requests.get(url)
res.raise_for_status()

df = pd.read_csv(io.StringIO(res.text), comment='#', header=None)
df.columns = ['Year', 'Month', 'Day', 'RMM1', 'RMM2', 'Amplitude', 'Phase', 'Dummy1', 'Dummy2']
df['Date'] = pd.to_datetime(df[['Year', 'Month', 'Day']])

# Save only RMM1
df[['Date', 'RMM1']].to_csv("RMM1.csv", index=False)

print("RMM1 Saved：RMM1.csv")
```
